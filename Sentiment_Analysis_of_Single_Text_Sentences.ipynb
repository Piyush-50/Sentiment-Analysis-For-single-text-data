## Stopword Removal

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))

stop_words.remove('not')

stop_words
```

```python

def remove_stopwords(text):
    return " ".join([word for word in str(text).split() if word not in stop_words])

df['text'] = df['text'].apply(lambda x : remove_stopwords(x))
test['text'] = test['text'].apply(lambda x: remove_stopwords(x))
df.head(5)
```

## Tokenisation

```python
from nltk.tokenize import sent_tokenize, word_tokenize

df['tokenized'] = df.apply(lambda row: nltk.word_tokenize(row['text']), axis=1)


df.head()
```

```python
test['tokenized']= test.apply(lambda x: nltk.word_tokenize(x['text']),axis=1)
test.head()
```

## Lemmatization


```python
lemmetizer = nltk.WordNetLemmatizer()

def lemmatize_txt(token_txt):
  data = [lemmetizer.lemmatize(word) for word in token_txt]
  
  return data
```

```python
df['lemmetized'] = df['tokenized'].apply (lambda a: lemmatize_txt(a))
df
```

```python
test['lemmetized'] = test['tokenized'].apply (lambda a: lemmatize_txt(a))
test
```

## Untokenize

```python
def list_to_string(a):
  st = " "
  return (st.join([b for b in a]))

df['lemmetized'] = df['lemmetized'].apply( lambda c: list_to_string(c))

```
